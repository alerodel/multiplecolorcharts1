```python
%matplotlib inline
#import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,100,1000)
y = np.exp(x/5)

fig = plt.figure(figsize=(9,9))

ax1 = fig.add_subplot(221) 
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223) 
ax4 = fig.add_subplot(224)

ax1.plot(x,y,lw=2, color = 'Blue')
ax2.plot(x,y,lw=2, color = 'Pink')
ax3.plot(x,y,lw=2, color = 'Red')
ax4.plot(x,y,lw=2, color = 'Yellow')

ax1.set_title('Azul')
ax2.set_title('Rosa')
ax3.set_title('Vermelho')
ax4.set_title('Amarelo')

ax1.set_ylabel('Eixo Y 1')
ax2.set_ylabel('Eixo Y 2')
ax3.set_ylabel('Eixo Y 3')
ax4.set_ylabel('Eixo Y 4')


ax1.set_xlabel('Eixo X 1')
ax2.set_xlabel('Eixo X 2')
ax3.set_xlabel('Eixo X 3')
ax4.set_xlabel('Eixo X 4')

ax1.grid(color='black',lw= 0.2)
ax2.grid(color='black',lw= 0.2)
ax3.grid(color='black',lw= 0.2)
ax4.grid(color='black',lw= 0.2)
```
